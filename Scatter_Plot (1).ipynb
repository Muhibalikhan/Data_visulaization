## Introduction to Scatter Plots
A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph. Each point represents the values of two variables, with one variable on the x-axis and the other on the y-axis. Scatter plots are useful for identifying relationships, patterns, or trends between two variables.

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Worldcup_channels_data.csv")
data.head(5)
```

```python
# Selecting relevant columns
selected_columns = ['Comment Count', 'Like count', 'Video Duration', 'View Count']
selected_data = data[selected_columns]

# Calculating the correlation matrix
correlation_matrix = selected_data.corr()

correlation_matrix
```

**Let's create scatter plot for view and like count**

```python
# Extracting relevant columns
view_count = data['View Count']
like_count = data['Like count']
```

```python
# Creating a scatter plot
plt.figure(figsize=(5, 3))
plt.scatter(view_count, like_count)
plt.show()
```

**X-tick and y-ticks**

**Title and x-y labels**

```python
plt.figure(figsize=(8, 3))
plt.scatter(view_count, like_count)

# Set labels and title
plt.title('View Count vs. Like Count')
#plt.title('View Count vs. Like Count',loc="right")
#plt.title('View Count vs. Like Count',loc="center",pad=100)
#plt.title('View Count vs. Like Count',color='Indigo')
plt.title('View Count vs. Like Count',color='Indigo',fontsize=12)


plt.xlabel('View Count')
#plt.xlabel('View Count',loc="center",labelpad=50)
plt.xlabel('View Count',loc="center",labelpad=50,color='teal')


plt.ylabel('Like Count')
#plt.ylabel('Like Count',loc="center")
#plt.ylabel('Like Count',loc="center",labelpad=50)
plt.ylabel('Like Count',loc="center",labelpad=50,color='green')




plt.xticks(np.linspace(min(view_count),max(view_count),10), rotation=45)

plt.show()
```

**Annotation**

```python
view_count = data['View Count']
like_count = data['Like count']


plt.figure(figsize=(8, 3))
plt.scatter(view_count, like_count,marker='*', color='green',label='Like Count')


plt.title('View Count vs. Like Count')
plt.xlabel('View Count')
plt.ylabel('Like Count')

plt.xticks(np.linspace(min(view_count),max(view_count),10), rotation=45)

#Find indices of the highest and minimum values
max_index = np.argmax(like_count)
min_index = np.argmin(like_count)

# Annotate the highest and minimum values
plt.annotate(f'{like_count.iloc[max_index]}', 
             (view_count.iloc[max_index], like_count.iloc[max_index]),
             textcoords="offset points", xytext=(0, 20), ha='center', color='red',fontsize=12, weight='bold')

plt.annotate(f'{like_count.iloc[min_index]}', 
             (view_count.iloc[min_index], like_count.iloc[min_index]),
             textcoords="offset points", xytext=(0, 20), ha='center', color='blue',fontsize=12, weight='bold')

plt.legend()
plt.grid(True)
plt.show()
```

**Plot multiple variables**

```python
view_count = data['View Count']
like_count = data['Like count']
comment_count = data['Comment Count']
video_duration = data['Video Duration']

plt.figure(figsize=(10, 6))
plt.scatter(view_count, like_count,marker='*', color='green',label='Like Count')
plt.scatter(view_count, comment_count,marker='o', color='orange',label='Comment Count')
plt.scatter(view_count, video_duration,marker='h', color='indigo',label='Video Duration')


plt.title('View Count vs. Like, comment and duration')
plt.xlabel('View Count')
#plt.ylabel('Like Count')

plt.xticks(np.linspace(min(view_count),max(view_count),10), rotation=45)

plt.legend()
plt.show()
```

```python
view_count = data['View Count']
like_count = data['Like count']
comment_count = data['Comment Count']
video_duration = data['Video Duration']

plt.figure(figsize=(10, 6))
plt.scatter(view_count, like_count,marker='*', color='green',label='Like Count')
plt.scatter(view_count, comment_count,marker='o', color='orange',label='Comment Count')
plt.scatter(view_count, video_duration,marker='h', color='indigo',label='Video Duration')


plt.title('View Count vs. Like, comment and duration')
plt.xlabel('View Count')
#plt.ylabel('Like Count')

plt.xticks(np.linspace(min(view_count),max(view_count),10), rotation=45)

plt.legend()
# Change the legend label color
legend = plt.legend()
legend.get_texts()[0].set_color("green")  # Change the color of the first legend label
legend.get_texts()[1].set_color("orange")  # Change the color of the second legend label
legend.get_texts()[2].set_color("indigo")  # Change the color of the second legend label

plt.grid(True)
plt.show()
```

**Engagement of different channels**

```python

# Creating a scatter plot with color-coded channels
plt.figure(figsize=(10, 8))

# Iterate through unique channels and plot data points with different colors
for channel in data['Channel Name'].unique():
    channel_data = data[data['Channel Name'] == channel]
    plt.scatter(channel_data['View Count'], channel_data['Like count'], label=channel, alpha=0.7)

plt.title('View Count vs. Like Count by Channel')
plt.xlabel('View Count')
plt.ylabel('Like Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python

from pandas.plotting import scatter_matrix

# Selecting relevant columns
selected_columns = ['Comment Count', 'Like count', 'Video Duration', 'View Count']
selected_data = data[selected_columns]

# Creating a scatter matrix plot
scatter_matrix(selected_data, alpha=0.7, figsize=(12, 10), diagonal='hist')

plt.suptitle('Scatter Matrix: Comment Count, Like Count, Video Duration, and View Count', y=0.92)
plt.show()
```
