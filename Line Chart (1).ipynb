## Introduction to Line Charts
A line chart is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Line charts are used to represent data over a continuous time span or to show trends.

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Worldcup_channels_data.csv")
data.head(5)
```

```python
data.shape
```

```python
data["Channel Name"].value_counts()
```

```python
#store the channel data in separate variable
har_lamha_purjosh_data = data[data['Channel Name'] == 'Har Lamha Purjosh']

# Group by 'Published Date' and sum 'View Count'
daily_progress = har_lamha_purjosh_data.groupby('Published Date')['View Count'].sum().reset_index()
daily_progress.head()
```

```python
plt.figure(figsize=(10, 3))
plt.plot(daily_progress['Published Date'], daily_progress['View Count'],marker='o',label='Summed View Count')

# Set labels and title
plt.title('Har Lamha Purjosh Channel - Summed View Count Over Time')
plt.xlabel('Published Date')

plt.ylabel('Summed View Count')
plt.xticks(daily_progress['Published Date'][::7], rotation=45)
plt.legend()
plt.grid(True)


# Annotate every 7th marker point with its 'View Count' value
for i, (date, view_count) in enumerate(zip(daily_progress['Published Date'], daily_progress['View Count'])):
    if i % 7 == 0:  # Annotate every 7th marker
        plt.annotate(f'{view_count:,}', (date, view_count))
      #  plt.annotate(f'{view_count:,}', (date, view_count), textcoords="offset points", xytext=(0, 10), ha='center')

plt.show()
```

```python
#store the channel data in separate variable
har_lamha_purjosh_data = data[data['Channel Name'] == 'Har Lamha Purjosh']

# Group by 'Published Date' and sum 'View Count'
daily_progress_view = har_lamha_purjosh_data.groupby('Published Date')['View Count'].sum().reset_index()

daily_progress_like = har_lamha_purjosh_data.groupby('Published Date')["Like count"].sum().reset_index()
```

```python
plt.figure(figsize=(10, 3))
plt.plot(daily_progress_view['Published Date'], daily_progress_view['View Count'],marker='o',label='Summed View Count')
plt.plot(daily_progress_like['Published Date'], daily_progress_like['Like count'],marker='o',label='Summed Like Count')

# Set labels and title
plt.title('Har Lamha Purjosh Channel - Engagement Over Time')
plt.xlabel('Published Date')

#plt.ylabel('Summed View Count')
plt.xticks(daily_progress['Published Date'][::7], rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

**Progress of different channels over single plot**

```python
data['Channel Name'].unique()
```

```python
# Get unique channel names
unique_channels = data['Channel Name'].unique()

# Dictionary to store daily progress DataFrames for each channel
channel_progress_dict = {}

# Iterate through each unique channel
for channel in unique_channels:
    # Store the channel data in a separate variable
    channel_data = data[data['Channel Name'] == channel]
    
    # Group by 'Published Date' and sum 'View Count'
    channel_daily_progress = channel_data.groupby('Published Date')['View Count'].sum().reset_index()
    
    # Store the channel's daily progress DataFrame in the dictionary
    channel_progress_dict[channel] = channel_daily_progress
```

```python
plt.figure(figsize=(12, 8))
plt.plot(channel_progress_dict['Har Lamha Purjosh']['Published Date'], channel_progress_dict['Har Lamha Purjosh']['View Count'],label='Har Lamha Purjosh')
plt.plot(channel_progress_dict['SUNO NEWS HD']['Published Date'], channel_progress_dict['SUNO NEWS HD']['View Count'],label='SUNO NEWS HD')
plt.plot(channel_progress_dict['GEO SUPER']['Published Date'], channel_progress_dict['GEO SUPER']['View Count'],label='GEO SUPER')
plt.plot(channel_progress_dict['A Sports']['Published Date'], channel_progress_dict['A Sports']['View Count'],label='A Sports')
plt.plot(channel_progress_dict['SAMAA TV']['Published Date'], channel_progress_dict['SAMAA TV']['View Count'],label='SAMAA TV')
plt.plot(channel_progress_dict['Hasna Mana Hai']['Published Date'], channel_progress_dict['Hasna Mana Hai']['View Count'],label='Hasna Mana Hai')

# Set labels and title
plt.title('Channel - Summed View Count Over Time')
plt.xlabel('Published Date')

plt.ylabel('Summed View Count')
plt.xticks(daily_progress['Published Date'][::7], rotation=45)
plt.legend()
plt.show()
```

le6 is a shorthand notation for 1 million (1,000,000) in Python. The 'le' stands for "lower or equal," and the '6' represents the power of 10.

**Sub plots**

```python
plt.figure(figsize=(12, 10))

# Iterate through each channel and plot on a separate subplot
for i, channel in enumerate(channel_progress_dict):
    plt.subplot(3, 2, i+1)
    plt.plot(channel_progress_dict[channel]['Published Date'], channel_progress_dict[channel]['View Count'], label=channel)
    plt.title(channel)

    # Turn off x and y ticks
    plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig("last_plot.png")
plt.show()
```
